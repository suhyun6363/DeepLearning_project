```python
import matplotlib.pyplot as plt
from matplotlib.image import imread
import cv2
import numpy as np
```

```python
img = imread('img_yn4.png')
plt.figure(figsize=(15,12))
plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(15,12))
plt.imshow(img_gray)
```

```python
img_blur = cv2.GaussianBlur(img_gray, (5, 5), 0)
plt.figure(figsize=(15,12))
plt.imshow(img_blur)
```

```python
#     
ret, img_th = cv2.threshold(img_blur, 0, 230, cv2.THRESH_BINARY_INV)
img_th = img_th.astype(np.uint8)
img_th
```

```python
contours, hierachy = cv2.findContours(img_th.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print(len(contours))
rects = [cv2.boundingRect(each) for each in contours]
rects
```
